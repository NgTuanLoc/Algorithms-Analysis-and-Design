# ABSTRACTION

Tìm số cặp ký tự liền kề giống nhau ở 2 chuỗi ký tự.

# PATTERN RECOGNITION
- Restriction: 
    + Time: 0.2s
    + Space: 50MB
    => Time Complexity Worst Case: 0.2 \* 10^8 = 2 * 10^7

Từ các giả thiết và yêu cầu của đề bài, ta cần thực hiện 2 bài toán sau:

- Tách chuỗi thành mảng chứa các cặp ký tự liền kề. => Có thể **vét cạn** vì độ dài chuỗi <10^5 và Time Restriction là 0.2s (O(n) với n là độ dài lớn nhất của 2 chuỗi).
- Tìm kiếm và đếm số phần tử giống nhau ở 2 mảng. 
    + Chuỗi tạo từ 26 ký tự => Mỗi mảng chứa các cặp liền kề có nhiều nhất 26\*26 = 676 cặp riêng biệt.
    + Vậy có thể vét cạn 2 vòng lặp vì O(676\*676) < Time Restriction.
    
=> Brute Force bằng cách lặp.

Worst Case Time Complexity: O(10^5)


# ALGORITHM DESIGN

- Ở bài toán tách chuỗi thành mảng chứa các cặp ký tự liền kề, để thuận tiện cho việc tìm kiếm và so sánh => Sử dụng hash containers như set hoặc dictionary. Nhóm chọn dictionary vì phù hợp để lưu keys là các cặp chuỗi liền kề, còn values sẽ là tần suất xuất hiện ở chuỗi input.

- Input: gen1, gen2 - lần lượt là chuỗi gene 1 và gene 2. (1< len(gen1), len(gen2) <=10^5)
- Output: Tổng các cặp ký tự liền kề giống nhau ở 2 chuỗi từ input.
- Temp:
    + Hàm dictionary: biến đổi 1 chuỗi thành 1 dict lưu keys là các cặp ký tự liền kề và values là tần suất xuất hiện của các keys trong chuỗi đó và trả về dict vừa tạo thành.
    + gen: tham số của hàm dictionary, nhận giá trị từ chuỗi được truyền vào.
    + dic: biến dạng dic chứa giá trị trả về từ hàm dictionary. (Initialize: dic = {})
    + dicA, dicB: biến dạng dict lưu keys là các cặp ký tự liền kề và values là tần suất xuất hiện của các keys trong gen1 và gen2. (Initialize: dicA = dictionary(gen1), dicB = dictionary(gen2))
    + sum: Biến kết quả output. (Initialize: sum = 0)
    + i: Biến chạy trong vòng lặp (Initialize: i = 0)

Các bước thực hiện:

+ Nhập gen1 và gen2 và khởi tạo sum = 0.
+ Hàm dictionary:
    + Khởi tạo dic rỗng.
    + Với i từ 0 đến len(gen)-1:
        + Nếu chuỗi gen\[i:i+2] chưa có trong dic -> Thêm phần tử có key là chuỗi gen[i:i+2] và value là 1.
        + Ngược lại -> value của key là chuỗi gen[i:i+2] =+1
    + Trả về dic.
+ Truyền giá trị trả về của dictionary(gen1) và dictionary(gen2) lần lượt cho dicA và dicB.
+ Với i và các phần tử trong dicB:
    + Nếu i có trong dicA:
        + sum += value của phần tử có key == i trong dicA.
+ Xuất output là sum.

# IMPLEMENTATION

```python
gen1 = input()
gen2 = input()
sum = 0
# gen1 = "ABBACAB"
# gen2 = "BCABB"

def dictionary(gen):
  dic = {}
  for i in range(len(gen)-1):
    if gen[i: i+2] not in dic:
      dic[gen[i: i+2]] = 1
    else:
      dic[gen[i: i+2]] +=1

  return dic

dicA = dictionary(gen1)
dicB = dictionary(gen2)

for i in dicB:
  if i in dicA:
    sum += dicA[i]

print(sum)
```
