# ABSTRACTION

Tm s cp k t lin k ging nhau  2 chui k t.

# PATTERN RECOGNITION
- Restriction: 
    + Time: 0.2s
    + Space: 50MB
    => Time Complexity Worst Case: 0.2 \* 10^8 = 2 * 10^7

T cc gi thit v yu cu ca  bi, ta cn thc hin 2 bi ton sau:

- Tch chui thnh mng cha cc cp k t lin k. => C th **vt cn** v  di chui <10^5 v Time Restriction l 0.2s (O(n) vi n l  di ln nht ca 2 chui).
- Tm kim v m s phn t ging nhau  2 mng. 
    + Chui to t 26 k t => Mi mng cha cc cp lin k c nhiu nht 26\*26 = 676 cp ring bit.
    + Vy c th vt cn 2 vng lp v O(676\*676) < Time Restriction.
    
=> Brute Force bng cch lp.

Worst Case Time Complexity: O(10^5)


# ALGORITHM DESIGN

-  bi ton tch chui thnh mng cha cc cp k t lin k,  thun tin cho vic tm kim v so snh => S dng hash containers nh set hoc dictionary. Nhm chn dictionary v ph hp  lu keys l cc cp chui lin k, cn values s l tn sut xut hin  chui input.

- Input: gen1, gen2 - ln lt l chui gene 1 v gene 2. (1< len(gen1), len(gen2) <=10^5)
- Output: Tng cc cp k t lin k ging nhau  2 chui t input.
- Temp:
    + Hm dictionary: bin i 1 chui thnh 1 dict lu keys l cc cp k t lin k v values l tn sut xut hin ca cc keys trong chui  v tr v dict va to thnh.
    + gen: tham s ca hm dictionary, nhn gi tr t chui c truyn vo.
    + dic: bin dng dic cha gi tr tr v t hm dictionary. (Initialize: dic = {})
    + dicA, dicB: bin dng dict lu keys l cc cp k t lin k v values l tn sut xut hin ca cc keys trong gen1 v gen2. (Initialize: dicA = dictionary(gen1), dicB = dictionary(gen2))
    + sum: Bin kt qu output. (Initialize: sum = 0)
    + i: Bin chy trong vng lp (Initialize: i = 0)

Cc bc thc hin:

+ Nhp gen1 v gen2 v khi to sum = 0.
+ Hm dictionary:
    + Khi to dic rng.
    + Vi i t 0 n len(gen)-1:
        + Nu chui gen\[i:i+2] cha c trong dic -> Thm phn t c key l chui gen[i:i+2] v value l 1.
        + Ngc li -> value ca key l chui gen[i:i+2] =+1
    + Tr v dic.
+ Truyn gi tr tr v ca dictionary(gen1) v dictionary(gen2) ln lt cho dicA v dicB.
+ Vi i v cc phn t trong dicB:
    + Nu i c trong dicA:
        + sum += value ca phn t c key == i trong dicA.
+ Xut output l sum.

# IMPLEMENTATION

```python
gen1 = input()
gen2 = input()
sum = 0
# gen1 = "ABBACAB"
# gen2 = "BCABB"

def dictionary(gen):
  dic = {}
  for i in range(len(gen)-1):
    if gen[i: i+2] not in dic:
      dic[gen[i: i+2]] = 1
    else:
      dic[gen[i: i+2]] +=1

  return dic

dicA = dictionary(gen1)
dicB = dictionary(gen2)

for i in dicB:
  if i in dicA:
    sum += dicA[i]

print(sum)
```
