<p>=>Khi chn s, ta cn chn s "dung ho" hai tham s ny  c th to c tng ln nht => Khong cch ca 2 tham s ny cng ln th tng t c cng ln</p>
<p>V vy ta s chn s 7 (v 7-3= 4 ln nht trong cc s)</p>
<p>- Khi chn s 7 vi s bng n cn l 3 ta c 3 trng hp sau</p>

    <li>Khng cn d bng n no (n%3=0)</li>

<h1>ALGORITHM DESIGN</h1>
<p><b>- Input:</b> S nguyn n (2<=n<= 10^6)</p>
<p>
    <b>-Temp</b>
    <ul>
        <li>n: S nguyn n nhp vo</li>
        <li>result: Kt qa</li>
        <li>case: Cc trng khi chia cho 3</li>
    </ul>
</p>
<p>
    <b>- Cc bc thc hin</b>
    <ol>
        <li> Nhp vo s nguyn n hp l</li>
        <li> result = n//3 (vi tng s bng n, chn ht s 7 vi s bng cn l 3)</li>
        <li> case = n %3 (Kim tra xem thuc trng hp no)</li>
        <ul>
            <li>case = 0: result * 7</li>
            <li>case = 1: (result -1)* 7 + 4</li>
            <li>case = 2: result * 7 + 1</li>            
        </ul>
        <li> Output:  s nguyn m - t tng ln nht</li>
    </ol>
</p>

```python
n = int(input())
result = n // 3
case = n % 3


if(case == 0):
    print(result * 7)
elif(case == 1):
    print((result - 1) * 7 + 4)
elif(case == 2):
    print(result * 7 + 1)
```
