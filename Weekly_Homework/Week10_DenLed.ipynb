<p><b>Bài toán 2: Tìm quy luật chọn số sao cho đạt tổng lớn nhất</b></p>
<p>- Sau khi tìm được quy luật của việc chọn số, ta nhận thấy có 2 tham số quan trọng là chọn số và số bóng đèn cần để tạo nên nó</p>
<p>- với số lượng bóng đèn cho trước: </p>
<ul>
    <li>Chọn số có <b>giá trị càng lớn</b> thì tổng cộng lại càng lớn</li>
    <li>Chọn số cần <b>càng ít bóng đèn </b>thì ta càng có nhiều cơ hội để chọn thêm số để tổng cộng lại càng lớn</li> </ul>
<p>=>Khi chọn số, ta cần chọn số "dung hoà" hai tham số này để có thể tạo được tổng lớn nhất => Khoảng cách của 2 tham số này càng lớn thì tổng đạt được càng lớn</p>
<p>Vì vậy ta sẽ chọn số 7 (vì 7-3= 4 lớn nhất trong các số)</p>
<p>- Khi chọn số 7 với số bóng đèn cần là 3 ta có 3 trường hợp sau</p>
<ul>
    <li>Không còn dư bóng đèn nào (n%3=0)</li>
    <li>Dư 1 bóng đèn (n%3=1): bỏ một số 7, chọn 4</li>
    <li>Dư 2 bóng đèn (n%3=2): chọn số 1</li>    
</ul>

<h1>ALGORITHM DESIGN</h1>
<p><b>- Input:</b> Số nguyên n (2<=n<= 10^6)</p>
<p>
    <b>-Temp</b>
    <ul>
        <li>n: Số nguyên n nhập vào</li>
        <li>result: Kết qủa</li>
        <li>case: Các trường khi chia cho 3</li>
    </ul>
</p>
<p>
    <b>- Các bước thực hiện</b>
    <ol>
        <li> Nhập vào số nguyên n hợp lệ</li>
        <li> result = n//3 (với tổng số bóng đèn, chọn hết số 7 với số bóng cần là 3)</li>
        <li> case = n %3 (Kiểm tra xem thuộc trường họp nào)</li>
        <ul>
            <li>case = 0: result * 7</li>
            <li>case = 1: (result -1)* 7 + 4</li>
            <li>case = 2: result * 7 + 1</li>            
        </ul>
        <li> Output:  số nguyên m - đạt tổng lớn nhất</li>
    </ol>
</p>

```python
n = int(input())
result = n // 3
case = n % 3


if(case == 0):
    print(result * 7)
elif(case == 1):
    print((result - 1) * 7 + 4)
elif(case == 2):
    print(result * 7 + 1)
```
