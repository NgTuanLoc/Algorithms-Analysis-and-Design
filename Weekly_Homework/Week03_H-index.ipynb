
# ABSTRACTION

Tìm length của subarray có giá trị min lớn hơn hoặc bằng số phần tử của subarray đó.


# PARTERN RECOGNITION

- Brute Force
- [1] Tìm độ dài mảng con chứa các bài báo có số lần trích dẫn không nhỏ hơn độ dài của mảng con này. 
- [2] Cần sắp xếp lại mảng chỉ số trích dẫn theo thứ tự tăng dần. Sau đó duyệt từ đầu đến hết mảng này để cập nhật H-index nếu:
    + chỉ số trích dẫn của bài báo đó lớn hơn hoặc bằng độ dài của subarray.
    + thỏa 1 trong 2 điều kiện:
        + vị trí của bài báo đang xét trong mảng < 1 (Để phòng trường hợp bài báo đầu mảng hay bài báo có số lượt trích dẫn nhỏ nhất lại được trích dẫn nhiều hơn length của subarray).
        + chỉ số trích dẫn ở bài báo liền trước không lớn hơn độ dài của subarray (để H_index được tiếp tục cập nhật nếu không thỏa điều kiện đầu).

- Độ phức tạp: 
    + [1]: O(nlogn)
    + [2]: O(n)
=> Độ phức tạp tiệm cận toàn bài toán: O(nlogn)

# ALGORITHM DESGIN

    Input: 
        n (số nguyên): Số bài báo. (1 ≤ n ≤ 5×10^5)
        c[] (số nguyên): mảng chứa chỉ số trích dẫn của từng bài báo (0 ≤ c[i] ≤ 10^6, i = 1 ÷ n)
    Output:
        H_index (số nguyên): Chỉ số H_index.
    Temp:
        i: biến đếm (0 ≤ i ≤ n-1).
        temp (số nguyên): biến lưu độ dài (số phần tử) subarray ở mỗi vòng lặp. (initialize: temp=n-i)
        

```python
n = int(input())
c = list(map(int, input().split()))
c.sort() # Time: O(nlogn)
H_index = 0


for i in range(n): # Time: O(n)
    temp = n - i
    if c[i] >= temp and (i < 1 or c[i - 1] <= temp):
        H_index = max(H_index, temp)
    if H_index != 0:
        break

print(H_index)
```
