

# Abstraction

* Given q pair of substring, check in each pair if 2 substring are permutation of each other

# Pattern recognition

* Solve string matching problem using prefix sum:
  * Given string S is a set of alphabetic character, each can be assign a value from 1 to 26
  * Represent the string using an array called Road, for 0 < i <= S.length, Road[i] = Road[i-1] + 3**S[i-1]

# Algorithm Design
*Note: Algorithm represented using psuedocode*

* Input:
  * s: string
  * q: integer: number of pair of substring
  * a,b,c,d: integer:for each pair, (a,b) and (c,d) are (start, end) index of each substring

* Output:<br>
  * q line for each line:
    * YES if 2 substring are permutation of each other
    * NO if otherwise

**Data structure**:<br>
* dic: hash map (static): map each alphabetic character with according value for easy calculation<br>
* road: dynamic array (static): used for representing the string<br>

**Psuedocode**

```python
INPUT s
INPUT q
dic = hash_map()
for i, item in enumerate(string.ascii_lowercase):
  dic[item] = i

road = dynamic_array()
road.append(0)

for i from 0 to s.length():
  road.append(road[i] + 3**dic[s[i]])
for i from 0 to q:
  INPUT a,b,c,d
  if road[a-1]-road[b] == road[c-1]-road[d]:
    print 'YES'
  else:
    print 'NO'
```

**Algorithm Analysis**

* Time complexity
  * Building a map: O(n)
  * Matching each pair of pattern: O(1) => With q pair of patterns, time complexity would be O(q)

* Space complexity
  * O(n) for n is the length of given string
  
# IMPLEMENTATION

```python
import sys
import string

s = input().strip()
q = int(input().strip())
road=[0]
        
dic = {}
for i, item in enumerate(string.ascii_lowercase):
    dic[item] = i
    
for i, value in enumerate(s):
    road.append(road[i] + 3**dic[value])

for _ in range(q):
    a, b, c, d = map(int, input().strip().split())
    
    if road[a-1]-road[b] == road[c-1]-road[d]:
        print('YES')
    else:
        print('NO')
```
